```python
import pandas as pd
import folium
```

```python
data1 = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
data1 = data1[data1["Borough"] != "Not assigned"]
data1
```

```python
data2 = pd.read_csv("Geospatial_data.csv")
data2
```

```python
data = pd.merge(data1, data2, on="Postal Code")
data
```

```python
m = folium.Map(location=[data["Latitude"].mean(), data["Longitude"].mean()], zoom_start=11)
for code, lat, lon in zip(data["Postal Code"], data["Latitude"], data["Longitude"]):
    folium.CircleMarker([lat, lon], radius=3, popup = code).add_to(m)
m
```
